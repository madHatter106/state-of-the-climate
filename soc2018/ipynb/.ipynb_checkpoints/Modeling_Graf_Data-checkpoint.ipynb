```python
import pickle
```

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as pl
from matplotlib import rcParams
from seaborn import PairGrid

import pymc3 as pm
```

```python
rcParams['axes.formatter.limits'] = (-2 ,3)
rcParams['axes.labelsize'] = 16
rcParams['xtick.labelsize'] = 16
rcParams['ytick.labelsize'] = 16
```

```python
datafile = '../carbon_data_from_jason.txt'
```

```python
df = pd.read_csv(datafile, delim_whitespace=True)
```

```python
df.head()
```

```python
f, ax = pl.subplots(figsize=(6, 6))
df.plot.scatter('bbp_470', 'phytoC', ax=ax)
ax.set_xlim(0, 0.004)
```

```python
df2 = pd.DataFrame(['EPO'] * 32 + ['AMT-22'] * 19, columns=['location'])
```

```python
df = pd.concat((df, df2),axis=1)
#df.append(pd.DataFrame(['EPO'] * 32 + ['AMT-22'] * 19, columns=['location']), sort=False)
```

```python
df.head()
```

```python
df['location_code'] = df.location.astype('category').cat.codes
```

```python
df.tail()
```

```python
bbp_470_std = df.bbp_470.std()
bbp_470_mean = df.bbp_470.mean()
df['bbp_470_s'] = (df.bbp_470 - bbp_470_mean) / bbp_470_std
```

```python
df.info()
```

```python
f, ax = pl.subplots(ncols=2, figsize=(12.5, 6))
df.loc[df.location=='EPO'].plot.scatter('bbp_470', 'phytoC', color='r', ax=ax[0])
df.loc[df.location!='EPO'].plot.scatter('bbp_470', 'phytoC', color='m', ax=ax[0])
df.loc[df.location=='EPO'].plot.scatter('bbp_470_s', 'phytoC', color='r', ax=ax[1])
df.loc[df.location!='EPO'].plot.scatter('bbp_470_s', 'phytoC', color='m', ax=ax[1])
ax[0].set_xlim(0, 0.004);
```

```python
with pm.Model() as m_pooled:
    h_icept_sd = pm.HalfCauchy('h_icept_sd', 2.5)
    h_slope_sd = pm.HalfCauchy('h_slope_sd', 2.5)
    icept_s = pm.Normal('icept_s', 0, h_icept_sd)
    slope_s = pm.Normal('slope_s', 0, h_slope_sd)
    sigma = pm.HalfCauchy('sigma', 5)
    mu_ = icept_s + df.bbp_470_s.values * slope_s
    y_lik = pm.Normal('likelihood', mu=mu_, sd=sigma, observed=df.phytoC.values)
    trace_pooled = pm.sample(2000, tune=2000)
```

```python
pm.model_to_graphviz(m_pooled)
```

```python
pm.traceplot(trace_pooled);
```

```python
with pm.Model() as m_pooled_2:
    h_icept_sd = pm.HalfCauchy('h_icept_sd', 10)
    h_slope_sd = pm.HalfCauchy('h_slope_sd', 10)
    icept_0 = pm.Normal('icept_0', 0, 1)
    slope_0 = pm.Normal('slope_0', 0, 1)
    icept = pm.Deterministic('icept', icept_0 * h_icept_sd)
    slope = pm.Deterministic('slope', slope_0 * h_slope_sd)
    sigma = pm.HalfCauchy('sigma', 5)
    mu_ = icept + df.bbp_470.values * 1000 * slope
    y_lik = pm.Normal('likelihood', mu=mu_, sd=sigma, observed=df.phytoC.values)
    trace_pooled_2 = pm.sample(2000, tune=2000, nuts_kwargs=dict(target_accept=0.9))
```

```python
pm.traceplot(trace_pooled_2);
```

```python
pm.plot_posterior(trace_pooled_2, varnames=['icept', 'slope']);
```

```python
with pm.Model() as m_hierarch:
    h_icept_sd = pm.HalfCauchy('h_icept', 2.5)
    h_slope_sd = pm.HalfCauchy('h_slope', 2.5)
    icept = pm.Normal('icept', 0, sd=h_icept_sd, shape=2)
    slope = pm.Normal('slope', 0, sd=h_slope_sd, shape=2)
    sigma = pm.HalfCauchy('sigma', 5)
    mu_ = icept[df.location_code.values] +\
            df.bbp_470_s.values * slope[df.location_code.values]
    y_lik = pm.Normal('likelihood', mu=mu_, sd=sigma, observed=df.phytoC.values)
```

```python
pm.model_to_graphviz(m_hierarch)
```

```python
with m_hierarch:
    trace_hierarch = pm.sample(2000, tune=2000)
```

```python
pm.traceplot(trace_hierarch);
```

```python
m_pooled.name = 'm_pooled'
m_pooled_2.name = 'm_pooled_uncentered'
m_hierarch.name = 'm_partial_pooled'
```

```python
compare_dict = {m_pooled: trace_pooled,
               m_pooled_2: trace_pooled_2,
               m_hierarch: trace_hierarch
               }
pm.compare(compare_dict)
```

```python
pm.compare(compare_dict, ic='LOO')
```
